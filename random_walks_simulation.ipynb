# Random Walk Simulation and Analysis

This notebook simulates random walks in 1D, 2D, and 3D dimensions, calculates the number of returns to the origin, and plots the trajectories and return distributions. 

We will:
- Simulate random walks in 1D, 2D, and 3D
- Analyze the number of times each walk returns to the origin
- Visualize the random walks and the distribution of returns

```python
import numpy as np
import matplotlib.pyplot as plt
from randomwalks import simulate_random_walk, plot_random_walks, plot_return_distribution
```

```python
# Parameters for simulation
n_steps = 5000
n_walks = 100000
dimensions = [1, 2, 3]

return_counts_1d = []
return_counts_2d = []
return_counts_3d = []
return_probabilities_1d = []
return_probabilities_2d = []
return_probabilities_3d = []

# Simulate random walks and analyze results
for dim in dimensions:
    walks, return_counts, return_probability = simulate_random_walk(dim, n_steps, n_walks)
    
    print(f"Dimension: {dim}D")
    print(f"Average number of returns to origin: {np.mean(return_counts)}")
    print(f"Probability of returning to the origin at least once: {return_probability:.4f}")
    print("-" * 40)
    
    # Store the return counts and probabilities
    if dim == 1:
        return_counts_1d = return_counts
        return_probabilities_1d = return_probability
    elif dim == 2:
        return_counts_2d = return_counts
        return_probabilities_2d = return_probability
    elif dim == 3:
        return_counts_3d = return_counts
        return_probabilities_3d = return_probability
    
    plot_random_walks(dim, walks)

# Plot the distribution of return counts
plot_return_distribution(return_counts_1d, return_counts_2d, return_counts_3d)
```
